```python
import spikeinterface.extractors as se
import spikeinterface.preprocessing as sp
import spikeinterface.core as sc
import spikeinterface.sortingcomponents.peak_detection as pd
import spikeinterface.sortingcomponents.tools as tools
import matplotlib.pyplot as plt
import numpy as np
import inspect

# Generate synthetic recording with proper scaling
recording = sc.generate_recording(
    durations=[10],               # 10 seconds
    sampling_frequency=30000,     # 30 kHz
    num_channels=4                # 4 channels
)
recording.set_channel_gains(1)
recording.set_channel_offsets(0)
print("Generated Recording:", recording)

# List available attributes for reference
print("Attributes in preprocessing:", dir(sp))
print("Attributes in peak_detection:", dir(pd))
print("Attributes in tools:", dir(tools))

# Perform peak detection using locally exclusive method
try:
    peaks = pd.detect_peaks(recording, method="locally_exclusive", detect_threshold=5, exclude_sweep_ms=0.5)
    print("Detected Peaks:", peaks[:5])  # Print first few detected peaks
except Exception as e:
    print("Error in peak detection:", e)

# Visualize data from channel 0
traces = recording.get_traces(segment_index=0)
plt.figure()
plt.plot(traces[:, 0])
plt.title("Sample Channel Data")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude (uV)")
plt.show()

# Extract noise levels and display
try:
    noise_levels = pd.get_noise_levels(recording)
    print("Noise Levels:", noise_levels)
except Exception as e:
    print("Error in noise level extraction:", e)

# Extract waveforms using ExtractSparseWaveforms
try:
    extractor = tools.ExtractSparseWaveforms(recording, ms_before=0.5, ms_after=0.5)
    waveforms = extractor.compute(traces, peaks)
    print("Extracted Waveforms Shape:", waveforms.shape)
except Exception as e:
    print("Error in waveform extraction:", e)

# Convert peaks to NumPy structured array if needed
peak_dtype = [('sample_index', 'int32'), 
              ('channel_index', 'int32'), 
              ('amplitude', 'float32'), 
              ('segment_index', 'int32')]

try:
    peaks_array = np.array(peaks, dtype=peak_dtype)
    waveforms = extractor.compute(traces, peaks_array)
    print("Extracted Waveforms Shape (from structured array):", waveforms.shape)
except Exception as e:
    print("Error using structured array for waveforms:", e)

# Inspect SortingAnalyzer methods
try:
    print(inspect.getmembers(sc.SortingAnalyzer, predicate=inspect.isfunction))
except Exception as e:
    print("Error inspecting SortingAnalyzer:", e)

# List extractors attributes
print("Attributes in extractors:", dir(se))
```

```python
import numpy as np

# Ensure peaks are within valid range
valid_peaks = []
num_samples = int(filtered_recording.get_total_duration() * filtered_recording.get_sampling_frequency())
ms_before = int(1 * filtered_recording.get_sampling_frequency() / 1000)  # 1 ms
ms_after = int(2 * filtered_recording.get_sampling_frequency() / 1000)   # 2 ms

for peak in peaks:
    if ms_before <= peak['sample_index'] < (num_samples - ms_after):
        valid_peaks.append(peak)

print(f"Valid Peaks Count: {len(valid_peaks)}")
```

```python
from spikeinterface.sortingcomponents.tools import ExtractSparseWaveforms

# Initialize extractor
extractor = ExtractSparseWaveforms(filtered_recording, ms_before=1, ms_after=2)

# Convert to NumPy structured array
peak_dtype = [('sample_index', 'int32'), 
              ('channel_index', 'int32'), 
              ('amplitude', 'float32'), 
              ('segment_index', 'int32')]
peaks_array = np.array(valid_peaks, dtype=peak_dtype)

# Extract waveforms
waveforms = extractor.compute(filtered_recording.get_traces(), peaks_array)

print("Extracted Waveforms Shape:", waveforms.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate

def align_by_cross_correlation(traces, sampling_frequency=30000, max_shift=200):
    """
    Aligns the signals from multiple channels based on cross-correlation.
    
    Parameters:
    traces (np.ndarray): A 2D array with shape (num_samples, num_channels).
    sampling_frequency (float): Sampling frequency of the signal.
    max_shift (int): Maximum shift in samples to consider for alignment.
    
    Returns:
    aligned_traces (np.ndarray): Aligned signal traces.
    shifts (list): List of shifts applied to each channel.
    """
    num_samples, num_channels = traces.shape
    aligned_traces = np.copy(traces)
    shifts = [0] * num_channels
    
    # Perform pairwise cross-correlation and find the optimal shifts
    for i in range(1, num_channels):
        # Cross-correlation between channel 0 and the current channel
        correlation = correlate(traces[:, 0], traces[:, i], mode='full')
        lags = np.arange(-len(traces[:, 0]) + 1, len(traces[:, 0]))
        
        # Find the lag (shift) that maximizes the correlation
        best_shift = lags[np.argmax(np.abs(correlation))]
        
        # Apply the shift
        if abs(best_shift) <= max_shift:
            aligned_traces[:, i] = np.roll(traces[:, i], -best_shift)
            shifts[i] = best_shift
        else:
            print(f"Skipping large shift ({best_shift}) for channel {i}")
    
    return aligned_traces, shifts
```
